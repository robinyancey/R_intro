**Conditions** allow your code to make decisions. In real life, we do this all the time:

Similarly, in R:

> ‚ÄúIf this condition is TRUE, do this; otherwise, do something else.‚Äù

##  **The `if` Statement**

````r
if (condition) {
  # Code to run if condition is TRUE
}
````

###  Example:

````r
x <- 5
if (x > 0) {
  print("x is positive")
}
````

**Explanation:**

* Since `x > 0` is `TRUE`, R runs the code inside the braces `{}`.

---

## **The `if...else` Statement**

````r
if (condition) {
  # Run if TRUE
} else {
  # Run if FALSE
}
````

###  Example:

````r
x <- -3
if (x > 0) {
  print("Positive number")
} else {
  print("Negative number or zero")
}
````

## ‚úÖ **The `if...else if...else` Chain**

````r
x <- 0
if (x > 0) {
  print("Positive")
} else if (x < 0) {
  print("Negative")
} else {
  print("Zero")
}
````

### Example usage of Logical oparators

Used when you **don‚Äôt know** how many times to loop ‚Äî just as long as the condition is TRUE.

Repeats **indefinitely** until a `break` statement is used.

````r
x <- 1
repeat {
  print(x)
  x <- x + 1
  if (x > 5) {
    break
  }
}
````

##  **Part 1: Exceptions (Error Handling)**

###  What is an Exception?

An **exception** (or error) occurs when your code runs into a problem it can't solve, like dividing by zero or accessing a missing variable.

###  Why Handle Exceptions?

* Prevent your script from crashing
* Provide meaningful messages
* Handle edge cases gracefully

---

###  Common Error Functions

* `stop()` ‚Äì Halts execution and throws an error
* `warning()` ‚Äì Shows a warning but continues execution
* `message()` ‚Äì Shows a message to the console (informational)
* `try()` ‚Äì Tries to run code; handles errors without crashing
* `tryCatch()` ‚Äì Advanced error handling, custom error logic

---

###  Example: `try()`

````r
bad_division <- function(x) {
  result <- try(10 / x, silent = TRUE)
  return(result)
}
bad_division(2)   # Returns 5
bad_division(0)
````

### Example: `tryCatch()`

````r
safe_divide <- function(a, b) {
  tryCatch({
    result <- a / b
    return(result)
  }, error = function(e) {
    return("Division error!")
  })
}
safe_divide(10, 0)
````

###Explanation
Why did it print Inf instead of "Division errror!"?

It prints "Inf" instead of "Division error!" because dividing by zero in R does not throw an error; it returns Inf (infinite) for numeric division:

### Why tryCatch() doesn't catch this:
The tryCatch() in the safe_divide() is only triggered by actual errors (i.e., runtime exceptions). But 10 / 0 is not an error in R - it's considered a valid operation with a result of Inf.

### How to Handle Division by Zero Properly
You need to explicitly check if b == 0 before doing the division.

````r
safe_divide <- function(a, b) {
  if (b == 0) {
    return("Division error!")
  }
  return(a / b)
}
````

````r
a <- safe_divide(10, 5)

b <- safe_divide(10, 0)

print(a)

print(b)
````

# Try and catch examples

````r
safe_sqrt <- function(x) {
  tryCatch({
    sqrt(x)
  }, warning = function(w) {
    cat("Warning: ", w$message, "\n")
    return(NA)
  }, error = function(e) {
    cat("Error: ", e$message, "\n")
    return(NA)
  }, finally = {
    cat("Finished attempting sqrt\n")
  })
}

safe_sqrt("hello")
safe_sqrt(-9)
safe_sqrt(9)
````

## **Fallback Value on Error**

````r
read_file_safe <- function(path) {
  tryCatch({
    readLines(path)
  }, error = function(e) {
    cat("Could not read file:", e$message, "\n")
    return(character(0))  # return empty character vector
  })
}

read_file_safe("nonexistent_file.txt")
````

##  Benchmarking with `microbenchmark`

You can use the **`microbenchmark`** package for more precise timing.

## What is microbenchmark()?
Microbenchmark function is a part of R's microbenchmark package, and it is used to accurately capture and compare the running time of short lengths of code.

It runs each expression some number of times (usually 100 by default), and gives fine-grained timing information for performance comparison.

````r
install.packages("microbenchmark")
library(microbenchmark)
````

````r
microbenchmark(
  sum_loop = {
    total <- 0
    for (x in rnorm(10000)) total <- total + x
  },
  sum_vec = sum(rnorm(10000))
)

# We see 200 rows and 2 columns
# The microbenchmark package tested the speed of sum_loop function and sum_vec function.
# It tested 100 times for each, that is why we see at the top 200 x 2
# we have 2 columns expr and time(double).
# in the first column is function name and second column time it took. (repeats 100 times default but we can change how many times it should repeat using times = n
# where n represents number of repetitions.)
````

## microbenchmart break down:

The code above compares the performance of two ways of summing 10,000 random numbers.

```r
total <- 0
for (x in rnorm(10000)) total <- total + x
```
* This generates 10,000 random normal numbers using rnorm(10000).

* Then it uses a for loop to manually add them one-by-one to a total variable.

* This is a slower, more manual way of summing.

```r
sum(rnorm(10000))
```
* This also generates 10,000 random normal numbers.

* It uses the built-in, vectorized sum() function to add them all at once.

* This is much faster due to R's internal optimizations for vectorized operations.

````r
# fewer repetition example.
microbenchmark(
  sum_loop = {
    total <- 0
    for (x in rnorm(10000)) total <- total + x
  },
  sum_vec = sum(rnorm(10000)),
  times = 10  # Only 10 evaluations
)
````

## 3.5 The Mean as a Least Squares Estimate

### What does this mean?

In statistics, **the mean** (average) of a set of numbers is **not just a summary** of the data - it's also a **solution to an optimization problem**. Specifically:

> The **mean** is the number that **minimizes** the **sum of squared deviations** (differences) from all data points.

This principle is called the **Least Squares Estimate**.

---

### Let's break it down:

Suppose you have a numeric dataset:

````r
x <- c(2, 4, 6, 8)
x
````

Each value in `x` can be compared to a central value (like the mean). The **deviation** of each value is how far it is from the mean:

* Deviation = `x[i] - mean(x)`

If you square each deviation (to make them all positive), you get the **squared deviation**.

Now, we can calculate the **sum of all squared deviations**:

````r
sum((x - mean(x))^2)
````

### Why is this useful?

Let's try to **minimize** this sum by trying different values for the ‚Äúcentral point‚Äù. The value that minimizes this sum will be the **best estimate of center**, using least squares logic.

In fact, **this minimum occurs exactly when the center is the mean**.

---

### Let's demonstrate it in R

#### Step 1: Create a dataset

````r
x <- c(2, 4, 6, 8, 10)
````

#### Step 2: Try a few values of ‚Äúcenter‚Äù

````r
# Try using 5, 6, and the actual mean
sum((x - 5)^2)
sum((x - 6)^2)
sum((x - mean(x))^2)  # This will be the smallest
````

#### Step 3: Generalize with a function

````r
squared_deviation_sum <- function(c, x) {
  sum((x - c)^2)
}

x <- c(2, 4, 6, 8, 10)
mean(x)  # Should be 6

squared_deviation_sum(5, x)      # Try 5
squared_deviation_sum(6, x)      # Try the mean
squared_deviation_sum(10, x)     # Try a value far from the mean
````

You will see that the **sum is smallest when c = mean(x)**.

---



### Visualization

If you have R's plotting capability:

````r
x <- c(2, 4, 6, 8, 10)
c_values <- seq(0, 12, 0.1)
errors <- sapply(c_values, function(c) sum((x - c)^2))

plot(c_values, errors, type = "l", col = "blue",
     xlab = "Candidate Mean (c)",
     ylab = "Sum of Squared Errors",
     main = "Minimum at the Mean")

abline(v = mean(x), col = "red", lty = 2)
````

This shows the sum of squared deviations is minimized at the mean.

---

### Why does this matter?

This idea forms the basis for:

* **Linear regression** (minimizing squared errors)
* **Statistical modeling**
* **Machine learning** loss functions
* **Signal processing**, and more

The mean is **not just a summary statistic**, it's the result of solving a **least squares problem**.

---

### Summary

* The **mean** of a set of numbers minimizes the **sum of squared deviations**.
* This is why it's called the **Least Squares Estimate**.
* This principle is foundational in statistics and modeling.
* In R, this can be demonstrated with simple code and plotting.

## 3.7 Non-linear Least Squares (NLS)

###  What is Least Squares?

The **least squares method** finds the best-fitting line or curve to a set of data points by **minimizing the sum of the squares of the vertical differences (residuals)** between the observed values and the model predictions.

###  Linear vs Non-linear Models

* **Linear least squares** is used when the relationship between variables is linear.
* **Non-linear least squares** is used when the relationship is **not linear**, meaning the equation contains variables raised to powers, exponential terms, logs, etc.

---

###  Example of a Non-linear Relationship

Suppose we believe the relationship between variables `x` and `y` follows this pattern:

````r
y = a * exp(b * x)
````

This is not a linear model because the parameter `b` is inside an exponent.

We will use R‚Äôs **`nls()`** function (non-linear least squares) to estimate `a` and `b` based on data.

---

###  `nls()` Function in R

````r
nls(formula, data, start)
````

* `formula`: a model like `y ~ a * exp(b * x)`
* `data`: a dataframe containing your x and y variables
* `start`: initial guesses for parameters (required!)

---

###  Step-by-Step Example in R

````r
# Sample data
x <- 1:10
y <- 5 * exp(0.3 * x) + rnorm(10, 0, 2)  # True a = 5, b = 0.3 + some noise

# Combine into a data frame
df <- data.frame(x, y)

# Fit a non-linear model: y = a * exp(b * x)
model <- nls(y ~ a * exp(b * x), data = df, start = list(a = 1, b = 0.1))

# View estimated parameters
summary(model)
````

###  Output (example):

This will return something like:

```
Parameters:
   Estimate Std. Error t value Pr(>|t|)    
a   5.128     0.401      12.8   <2e-16 ***
b   0.294     0.025      11.8   <2e-16 ***
```

 These are estimates of the parameters `a` and `b` that best fit the data.

---

###  Plotting the Fit

````r
# Plot actual data
plot(df$x, df$y, main = "Non-linear Least Squares Fit", xlab = "x", ylab = "y")

# Add the fitted curve
lines(df$x, predict(model), col = "red", lwd = 2)
````

###  Important Notes

1. **Non-linear models can be tricky** ‚Äì they might not converge (i.e., fail to find a solution) if your initial guesses (`start`) are way off.
2. You must **understand the shape of your data** to choose a suitable model.
3. Use **plotting** to verify that your model actually fits the data well.

---

###  When Do We Use Non-linear Least Squares?

* When modeling **growth curves** (e.g., bacteria population).
* In **pharmacokinetics** (how a drug concentration changes over time).
* In **economics** (e.g., utility functions).
* In **machine learning** (fitting non-linear functions).

---

###  Summary

* **`nls()`** is used to estimate parameters in **non-linear models**.
* You must **supply a model formula and starting values**.
* Plotting helps to **visualize model performance**.
* Always **check the output** to ensure convergence and parameter accuracy.

````r
# Simulated vector of numbers (some will cause errors on purpose)
numbers <- c(10, 20, "error", 40, NA, 60)

# Function to safely process each number
safe_process <- function(x) {
  tryCatch({
    num <- as.numeric(x)

    if (is.na(num)) {
      stop("Value is NA or not convertible to numeric.")
    }

    result <- num * 2
    message("‚úÖ Processed value: ", result)
    return(result)

  }, warning = function(w) {
    message("‚ö†Ô∏è Warning: ", w$message)
    return(NA)
  }, error = function(e) {
    message("‚ùå Error: ", e$message)
    return(NA)
  })
}

# Store results
results <- c()

# Safe loop
for (i in seq_along(numbers)) {
  message("üîÑ Processing index ", i)

  value <- numbers[i]

  # Explicit NA check first
  if (is.na(value)) {
    message("‚ö†Ô∏è Value is NA, breaking loop.")
    break
  }

  # Safe comparison
  if (identical(value, "error")) {
    message("‚è≠Ô∏è Skipping value 'error'")
    next
  }

  processed <- safe_process(value)

  if (is.na(processed)) {
    message("üõë Processed value is NA, breaking loop.")
    break
  }

  results <- c(results, processed)
}

# Final output
message("‚úÖ Final results:")
print(results)
````

* ### Handling text
#### Example

````r
###  Handling Text in R with `tryCatch`
# Function to safely clean and transform a string
safe_text_process <- function(text) {
  tryCatch({
    if (is.na(text) || text == "") stop("Empty or NA text input")

    # Basic transformations
    cleaned <- trimws(text)                  # Remove leading/trailing whitespace
    upper <- toupper(cleaned)                # Convert to uppercase
    replaced <- gsub("HELLO", "HI", upper)   # Replace words
    splitted <- strsplit(replaced, " ")[[1]] # Split into words

    message("‚úÖ Successfully processed: ", text)
    return(splitted)

  }, warning = function(w) {
    message("‚ö†Ô∏è Warning: ", w$message)
    return(NA)
  }, error = function(e) {
    message("‚ùå Error: ", e$message)
    return(NA)
  })
}

# Example character vector (some bad inputs)
texts <- c("  hello world  ", NA, "", "Hello R", "text-processing")

# Process each text
results <- list()

for (i in seq_along(texts)) {
  message("üîÑ Processing text ", i)
  processed <- safe_text_process(texts[i])
  results[[i]] <- processed
}

# View results
print(results)
````

````r
# Function to safely get an environment variable
safe_get_env <- function(varname) {
  tryCatch({
    value <- Sys.getenv(varname)
    if (value == "") stop(paste("Environment variable", varname, "not found"))
    message("‚úÖ Found environment variable: ", varname)
    return(value)
  }, error = function(e) {
    message("‚ùå Error: ", e$message)
    return(NULL)
  })
}
````

````r
# Function to safely run a system command
safe_system_cmd <- function(command) {
  tryCatch({
    message("‚öôÔ∏è Running command: ", command)
    output <- system(command, intern = TRUE)
    message("‚úÖ Command ran successfully")
    return(output)
  }, error = function(e) {
    message("‚ùå System command failed: ", e$message)
    return(NULL)
  })
}
````

````r
# Function to safely get working directory
safe_getwd <- function() {
  tryCatch({
    wd <- getwd()
    message("üìÅ Current working directory: ", wd)
    return(wd)
  }, error = function(e) {
    message("‚ùå Error getting working directory: ", e$message)
    return(NULL)
  })
}
````
