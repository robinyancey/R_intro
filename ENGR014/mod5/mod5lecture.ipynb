## **2. Practical Examples with Customization**  

### **Example 1: Scatterplot (`plot()`)**  
**Use case:** Relationship between height and weight.

````r
tryCatch({
  height <- c(150, 160, 165, 170, 175)
  weight <- c(55, 60, 65, 68, 75)

  plot(height, weight,
       main = "Height vs. Weight (Students)",
       xlab = "Height (cm)", ylab = "Weight (kg)",
       col = "darkblue", pch = 16, cex = 1.2)
  grid()  # Add gridlines for readability
}, error = function(e) {
  message("Scatterplot error: ", e$message)
})
````

## Explanation
```r
  height <- c(150, 160, 165, 170, 175)
  weight <- c(55, 60, 65, 68, 75)
```
* Defines two vectors
  * height: students' heights in centimeters
  * weight: students' weights in kilograms.
  * These will be used as x and y coordinates in the scatterplot

```r
  plot(height, weight,
       main = "Height vs. Weight (Students)",
       xlab = "Height (cm)", ylab = "Weight (kg)",
       col = "darkblue", pch = 16, cex = 1.2)
```

* Creates a scatterplot:

  * `height` on the x-axis, `weight` on the y-axis.

  * `main`: Title of the plot.

  * `xlab`, `ylab`: Axis labels.

  * `col`: Sets point color to dark blue.

  * `pch = 16`: Filled circles as points.

  * `cex = 1.2`: Slightly enlarges point size.

```r
  grid()  # Add gridlines for readability
```

### **Example 2: Barplot (`barplot()`)**  
**Use case:** Student enrollment by major.  

````r
tryCatch({
  majors <- c("Engineering", "Business", "Arts", "Sciences", "Law")
  counts <- c(120, 150, 90, 110, 30)

  barplot(counts,
          names.arg = majors,
          main = "Student Distribution by Major",
          col = c("skyblue", "salmon", "gold", "lightgreen", "plum"),
          ylab = "Number of Students",
          border = NA)  # Remove borders for cleaner look
}, error = function(e) {
  message("Barplot error: ", e$message)
})
````

## Explanations

```r
majors <- c("Engineering", "Business", "Arts", "Sciences", "Law")
counts <- c(120, 150, 90, 110, 30)
```
* `majors`: A vector of student majors (used as labels for each bar).

* `counts`: A vector with the number of students in each major.

* These two vectors must match in length.


```r
barplot(counts,
        names.arg = majors,
        main = "Student Distribution by Major",
        col = c("skyblue", "salmon", "gold", "lightgreen", "plum"),
        ylab = "Number of Students",
        border = NA)
```

* Creates a vertical bar chart:

counts: Heights of the bars.

  * `names.arg`: Labels under each bar (from majors).

  * `main`: Title of the chart.

  * `col`: Colors for each bar, using a different color per major.

  * `ylab`: Label for the y-axis.

  * `border = NA`: Removes borders around the bars for a cleaner look.

### **Example 4: Boxplot (`boxplot()`)**

### **Example 5: Line Plot (`plot()` + `lines()`)**

````r
plot(year, enrollment,

     xlab = "Year", ylab = "Students",
````


  * `main`, `xlab`, `ylab`: Plot title and axis labels.

  * `ylim = c(300, 550)`: Sets y-axis limits to ensure consistent visual scale.

```r
abline(h = 500, lty = 2, col = "red")
```

* Adds a horizontal reference line at `y = 500`:

  * `lty = 2`: Dashed line.

  * `col = "red"`: Makes it stand out — useful for visual threshold (e.g. target enrollment).

## **Visual Customization Techniques**

### 1. **Point Customization**

````r
# Advanced point customization example
set.seed(123)
x <- rnorm(50)
y <- 2*x + rnorm(50, sd=0.5)

plot(x, y,
     # Point shape: 4 different shapes for 4 groups (15 to 18 are square, diamond, etc.)
     pch = c(15:18)[as.factor(rep(1:4, length.out=50))],  # Different shapes
     # Point color: Rainbow palette for each group, semi-transparent using alpha
     col = rainbow(4, alpha=0.7)[as.factor(rep(1:4, length.out=50))],  # Transparent colors
     # point size: Sequence from size 1 to 2 across the 50 points.
     cex = seq(1, 2, length.out=50),  # Varying sizes
     main = "Advanced Point Customization",
     xlab = "X Variable",
     ylab = "Y Variable")
grid() # this adds background grid lines to the plot
legend("topleft",
legend=paste("Group", 1:4), # Labels like "Group 1", "Group 2", etc.
       pch=15:18, # matching point shapes
       col=rainbow(4), pt.cex=1.5 # size of points in legend
       )
````

## Explanations

- This ensures reproducibility of random values
```r
set.seed(123)
```
- Generates random x values from normal distribution
```r
x <- rnorm(50)
```
- Generates corresponding y values based on a linear relation with x + noise
```r
y <- 2 * x + rnorm(50, sd = 0.5)
```

## Visual Breakdown

| Feature      | Code Element                     | Meaning                               |
| ------------ | -------------------------------- | ------------------------------------- |
| **Shapes**   | `pch = 15:18`                    | Different shapes for each group       |
| **Colors**   | `col = rainbow(4, alpha=0.7)`    | 4 colors, slightly transparent        |
| **Sizes**    | `cex = seq(1, 2, length.out=50)` | Gradually increases point sizes       |
| **Grouping** | `rep(1:4, length.out = 50)`      | Assigns each point to one of 4 groups |
| **Legend**   | `legend("topleft", ...)`         | Describes groups with color/shape     |

### 1. **Regression Lines with Confidence Intervals**

````r
model <- lm(y ~ x)
newx <- seq(min(x), max(x), length.out=100)
preds <- predict(model, newdata=data.frame(x=newx), interval="confidence")

plot(x, y, main="Regression with Confidence Band")
abline(model, col="red", lwd=2)
matlines(newx, preds[, c("lwr", "upr")],
         col="blue", lty=2, lwd=1.5)
````

## Explanations
```r
model <- lm(y ~ x)
```

* Creates a linear model (regression line) of `y` as a function of `x`.

* `lm()` stands for "linear model".

* This line fits a straight line:

  `y = a + bx`

```r
newx <- seq(min(x), max(x), length.out = 100)
```
* Generates 100 evenly spaced `x` values between the min and max of the original data.

* These are used to smoothly draw the confidence interval curve.

```r
preds <- predict(model, newdata = data.frame(x = newx), interval = "confidence")
```

* Uses the regression model to predict `y` values (and their confidence intervals) for the new `x` values.

* `interval = "confidence"` returns a matrix with:

  * `fit`: predicted y values

  * `lwr`: lower bound of the confidence interval

  * `upr`: upper bound of the confidence interval

```r
plot(x, y, main = "Regression with Confidence Band")
```

* Plots the original data points (`x, y`) as a scatterplot.

* Adds the title: “Regression with Confidence Band”.

```r
abline(model, col = "red", lwd = 2)
```

* Adds the regression line to the plot.

* col = "red": line color.

* lwd = 2: line thickness.

```r
matlines(newx, preds[, c("lwr", "upr")],
col = "blue", lty = 2, lwd = 1.5)
```

* Adds the confidence band:

  * `matlines()` draws multiple lines at once.

  * `preds[, c("lwr", "upr")]`: selects the lower and upper confidence bounds.

  * `col = "blue"`: color of the bands.

  * `lty = 2`: dashed lines.

  * `lwd = 1.5`: line thickness.

#### *Visual Output:*
* A scatterplot of data points.

* A red regression line.

* Two dashed blue lines showing the 95% confidence interval around the prediction.



### 2. **Non-linear Smoothing**

##### To smooth the relationship between `x` and `y` using LOESS (Locally Estimated Scatterplot Smoothing) and plot it on a scatterplot

````r
# LOESS smoothing
plot(x, y, main="LOESS Smoothing")
lines(loess.smooth(x, y), col="purple", lwd=2)
````

## Explanations

```r
# LOESS smoothing
plot(x, y, main = "LOESS Smoothing")
```

* Plots the raw data points (`x, y`) as a scatterplot.

* Adds the title “LOESS Smoothing”.

* By default, `plot()` will draw the points.

```r
lines(loess.smooth(x, y), col = "purple", lwd = 2)
```

#### This line is key:

#### What it does:
* `loess.smooth(x, y)`:

  * Applies **LOESS** smoothing to your data.

  * Internally fits local regressions over the data range.

  * Returns a list with:

    * x: sorted input values

    * y: smoothed y-values

* `lines(...):`

  * Draws a smooth curve on top of the scatterplot.

  * `col` = "purple": makes the curve purple.

  * `lwd`= 2: makes the curve thicker than default.

### Result:
#### You get:
  * A scatterplot of your original data.

  * A smooth purple curve that follows the trend of the data, useful when the relationship is not perfectly linear.

## 5.2 Modifications of Histograms

**Topic Goal:**
Learn how to customize and interpret histograms, adjust bin widths, colors, breaks, and overlay helpful visual guides like density lines or normal curves.

---

### What is a Histogram?

A **histogram** is a graphical representation of the distribution of a numeric variable. It divides the range of the data into **intervals ("bins")** and counts how many observations fall into each interval.
Unlike a bar chart, a histogram is used for **continuous data** and bins are shown in order.

---

### Key Histogram Parameters in R

* `breaks`: Controls the number of bins or how bins are defined
* `col`: Specifies the fill color of the bars
* `border`: Changes the color of the bar borders
* `xlim`, `ylim`: Set the x- and y-axis ranges
* `main`, `xlab`, `ylab`: Plot title and axis labels
* `freq`: If `TRUE`, shows frequencies; if `FALSE`, shows probability densities

---

### Example 1: Basic Histogram with Color and Axis Labels

````r
# Create a basic histogram of normally distributed data
tryCatch({
  data <- rnorm(100, mean = 50, sd = 10)  # Generate 100 random values

  hist(data,
       main = "Histogram of Normally Distributed Data",
       xlab = "Values",
       ylab = "Frequency",
       col = "lightblue",
       border = "black")
}, error = function(e) {
  message("Error creating basic histogram:", e$message)
})
````

**Explanation:**

* `rnorm(100, mean = 50, sd = 10)`: Generates 100 random numbers from a normal distribution.
* `hist()` is used to create the histogram.

## 5.3 Parallel Box-Plots

**Topic Goal:**
Learn how to create and customize multiple box-plots side-by-side (parallel box-plots) to compare distributions of data across groups or categories.

---

### What is a Box-Plot?

A **box-plot** (or box-and-whisker plot) is a standardized way to display the distribution of data based on five summary statistics:

* Minimum
* First quartile (Q1)
* Median (Q2)
* Third quartile (Q3)
* Maximum

It also visually highlights **outliers** if present.

Box-plots are very useful for **comparing distributions across multiple categories** because they summarize shape, spread, and center of data concisely.

---

### Syntax for Box-Plots in R

```r
boxplot(formula, data, ...)
```

* `formula`: `value ~ group`
* `data`: A data frame containing the variables
* Additional options include `main`, `xlab`, `ylab`, `col`, `notch`, etc.

---

### Example 1: Basic Parallel Box-Plot Comparing Three Groups

````r
# Create parallel box-plots of exam scores by subject
tryCatch({
  scores <- data.frame(
    Subject = rep(c("Math", "Science", "History"), each = 20),
    Score = c(
      rnorm(20, mean = 80, sd = 5),
      rnorm(20, mean = 75, sd = 8),
      rnorm(20, mean = 70, sd = 6)
    )
  )

  boxplot(Score ~ Subject,
          data = scores,
          main = "Exam Scores by Subject",
          xlab = "Subject",
          ylab = "Score",
          col = c("lightblue", "lightgreen", "lightpink"))
}, error = function(e) {
  message("Error creating box-plots:", e$message)
})
````

**Explanation:**

* `rep(c(...), each = 20)`: Repeats the subject labels so that each group has 20 values.
* `rnorm(...)`: Generates random scores for each subject.
* `Score ~ Subject`: Formula syntax tells R to group Score by Subject.
* `col = c(...)`: Sets different fill colors for each box.
* This plot allows you to visually compare central tendency and spread of scores across subjects.

---

### Example 2: Adding Notches and Horizontal Orientation

````r
# Box-plots with notches and horizontal layout
tryCatch({
  boxplot(Score ~ Subject,
          data = scores,
          notch = FALSE,
          horizontal = TRUE,
          col = "wheat",
          main = "Horizontal Notched Box-Plot")
}, error = function(e) {
  message("Error in notched box-plot:", e$message)
})
````

**Explanation:**
* `horizontal = TRUE`: Rotates the box-plots horizontally for better readability in some cases.

---

### Example 3: Overlaying Mean Points on Box-Plots


````r
# Add mean values to each box
tryCatch({
  boxplot(Score ~ Subject,
          data = scores,
          col = "lavender",
          main = "Box-Plot with Mean Points")

  means <- tapply(scores$Score, scores$Subject, mean)
  points(1:3, means, col = "red", pch = 18, cex = 1.5)
}, error = function(e) {
  message("Error adding mean points:", e$message)
})
````

**Explanation:**

* `tapply(...)`: Calculates mean scores for each subject group.
* `points()`: Adds red diamond-shaped points (`pch = 18`) to represent the mean of each group.
* This is useful to compare median (from box) vs. mean (from point) visually.

---

### Best Practices for Parallel Box-Plots

* Always provide meaningful **axis labels** and **titles**.
* Use color to **differentiate categories**, but avoid overwhelming the viewer.
* Overlaying **mean points** helps when data is skewed and the mean is far from the median.
* Use **notches** to visually test if medians are significantly different (non-overlapping notches = likely difference).

## 5.4 Advanced Plot Customization

**Topic Goal:**
Learn how to refine and control plots using advanced graphical parameters in R. Customize every element of your visualizations: axes, legends, margins, text, and multi-plot layouts.

---

### Why Customize Plots?

Customizing plots is critical when:

* Preparing figures for reports, academic papers, or presentations.
* Making graphics accessible and clear for viewers.
* Displaying multiple visualizations in one view.

---

### Advanced Graphical Parameters with `par()`

The function `par()` in R is used to set or query graphical parameters. It is powerful and allows for extensive plot customization.

Some commonly used options include:

| Parameter                    | Description                           |
| ---------------------------- | ------------------------------------- |
| `mfrow`                      | Split screen into rows x columns      |
| `mar`                        | Margin size: bottom, left, top, right |
| `las`                        | Text orientation (0-3)                |
| `cex`, `cex.lab`, `cex.main` | Text size                             |
| `col.axis`, `col.lab`        | Axis and label colors                 |

---

### Example 1: Display Multiple Plots on the Same Page

````r
# Plot 4 different plots on a 2x2 grid
tryCatch({
  par(mfrow = c(2, 2))  # 2 rows, 2 columns layout

  plot(rnorm(50), main = "Plot 1")
  hist(rnorm(50), main = "Histogram")
  boxplot(rnorm(50), main = "Boxplot")
  barplot(table(sample(letters[1:5], 100, replace = TRUE)), main = "Barplot")

  par(mfrow = c(1, 1))  # Reset to default
}, error = function(e) {
  message("Error in multi-plot layout: ", e$message)
})
````

**Explanation:**

* `par(mfrow = c(2, 2))` splits the plotting area into 2 rows and 2 columns.
* Four plots are created in sequence and automatically arranged in this grid.

---

### Example 2: Customize Margins, Axis Text, and Titles

````r
# Customized plot with adjusted margins and axis style
tryCatch({
  par(mar = c(5, 5, 4, 2), las = 1, cex.lab = 1.2, cex.main = 1.5)

  x <- 1:10
  y <- x^2

  plot(x, y,
       type = "b",
       col = "darkblue",
       pch = 16,
       main = "Customized Quadratic Plot",
       xlab = "X Values",
       ylab = "Y = X^2")

  par(mar = c(5, 4, 4, 2))  # Reset margins
}, error = function(e) {
  message("Error customizing margins or labels: ", e$message)
})
````

**Explanation:**

* `mar = c(5, 5, 4, 2)`: Adjusts margin sizes (bottom, left, top, right).
* `las = 1`: Makes all axis labels horizontal.
* `cex.lab` and `cex.main`: Control font size of labels and title.
* `pch = 16`: Uses filled circles as points.
* `type = "b"`: Shows both points and lines.

---

### Example 3: Adding a Custom Legend to a Plot

````r
# Line plot with legend and multiple lines
tryCatch({
  x <- 1:10
  y1 <- x * 2
  y2 <- x * 3

  plot(x, y1, type = "l", col = "blue", lwd = 2,
       ylim = c(0, 35), ylab = "Value", xlab = "X",
       main = "Multiple Lines with Legend")
  lines(x, y2, col = "red", lwd = 2)

  legend("topleft",
         legend = c("y = 2x", "y = 3x"),
         col = c("blue", "red"),
         lty = 1,
         lwd = 2)
}, error = function(e) {
  message("Error adding legend or lines: ", e$message)
})
````

**Explanation:**

* `legend("topleft", ...)`: Adds a key to explain the lines.
* `col`, `lty`, and `lwd` ensure visual match between legend and lines.
* `lines()` is used to overlay multiple line graphs on one plot.

---

### Tips for Advanced Plot Customization

* Use `par()` to pre-set layout and styling before drawing plots.
* Always **reset `par()`** settings after customized plots to avoid affecting future plots.
* Consistent **line widths (`lwd`)** and point types (`pch`) improve readability.
* Add **legends** for clarity when plotting multiple lines or categories.


## Stacked Bar Chart: Degrees Awarded by Type of University

This example demonstrates how to generate a stacked bar chart using the columns: `CONTROL`, `CREDDESC`, and `INSTNM` from the dataset `Most-Recent-Cohorts-Field-of-Study.csv`. We group by `CONTROL` (institution type) and stack by `CREDDESC` (degree type).

We also set a fixed Y-axis scale from 0 to 140,000 in 20,000 intervals, and define a specific plot width and height.


````r
# Required: run this once to enable custom output sizes in Jupyter Notebook
options(repr.plot.width = 15, repr.plot.height = 5)  # width and height in inches

tryCatch({
  library(dplyr)
  library(ggplot2)
  library(readr)
  library(scales)

  # Load CSV
  df <- read_csv("Most-Recent-Cohorts-Field-of-Study.csv", show_col_types = FALSE)

  # Summarize by CONTROL and CREDDESC
  summary_df <- df %>%
    filter(!is.na(CONTROL), !is.na(CREDDESC)) %>%
    group_by(CONTROL, CREDDESC) %>%
    summarise(Count = n(), .groups = "drop")

  # Plot
  ggplot(summary_df, aes(x = CONTROL, y = Count, fill = CREDDESC)) +
    geom_bar(stat = "identity") +
    labs(title = "Degrees Awarded by Type of University",
         x = "Type of University",
         y = "Headcount",
         fill = "Degree Type") +
    scale_y_continuous(
      limits = c(0, 140000),
      breaks = seq(0, 140000, by = 20000),
      labels = comma
    ) +
    theme_minimal() +
    theme(
    axis.text.x = element_text(angle = 0, hjust = 0.5),
    plot.title = element_text(hjust = 0.5)  # Centers the title
    )
}, error = function(e) {
  message("Error creating plot: ", e$message)
})
````

### Percentage of Degree Types by University Type (100% Stacked Bar)

````r
tryCatch({
  library(ggplot2)
  library(dplyr)
  library(readr)

  df <- read_csv("Most-Recent-Cohorts-Field-of-Study.csv", show_col_types = FALSE)

  # Group and calculate percentages within CONTROL
  options(repr.plot.width = 15, repr.plot.height = 5)
  percent_df <- df %>%
    filter(!is.na(CONTROL), !is.na(CREDDESC)) %>%
    group_by(CONTROL, CREDDESC) %>%
    summarise(Count = n(), .groups = "drop") %>%
    group_by(CONTROL) %>%
    mutate(Percent = Count / sum(Count) * 100)

  ggplot(percent_df, aes(x = CONTROL, y = Percent, fill = CREDDESC)) +
    geom_bar(stat = "identity", position = "fill") +
    scale_y_continuous(labels = scales::percent) +
    labs(title = "Percentage of Degrees Awarded by Type of University",
         x = "Type of University",
         y = "Percent of Degrees Awarded",
         fill = "Degree Type") +
    theme_minimal() +
    theme(
      plot.title = element_text(hjust = 0.5, size = 14),
      axis.text.x = element_text(angle = 0, hjust = 0.5)
    )
}, error = function(e) {
  message("Error generating 100% stacked bar chart: ", e$message)
})
````

### Defining and Using Colors in R

In R, you can define colors using:

* **Named colors**: e.g. `"red"`, `"skyblue"`, `"darkgreen"` (see `colors()`)
* **Hex codes**: e.g. `"#FF0000"` (for red)
* **RGB values**: using `rgb()` function
* **Color palettes**: e.g. `rainbow()`, `heat.colors()`, `topo.colors()`

---

### Example 1: Using a Color Palette in a Plot

````r
# Barplot with rainbow colors
tryCatch({
  categories <- c("A", "B", "C", "D", "E")
  values <- c(10, 25, 30, 15, 20)

  barplot(values,
          names.arg = categories,
          col = rainbow(length(categories)),
          main = "Barplot with Rainbow Colors")
}, error = function(e) {
  message("Error in color palette barplot: ", e$message)
})
````

**Explanation:**

* `rainbow(n)` generates `n` distinct vibrant colors.
* Each bar is assigned a different color from the palette.

---

### Plotting in Higher Dimensions

“Higher dimensions” in data visualization refers to visualizing more than two variables simultaneously. You can represent extra dimensions through:

* **Color**
* **Point size**
* **Shape**
* **Transparency**
* **Panels/facets** (e.g. in `ggplot2`)

---

### Example 2: Scatterplot with Color Representing a Third Variable

````r
# Scatterplot where point color represents group
tryCatch({
  set.seed(123)
  x <- rnorm(50)
  y <- rnorm(50)
  group <- sample(c("Group 1", "Group 2"), 50, replace = TRUE)

  colors <- ifelse(group == "Group 1", "blue", "orange")

  plot(x, y,
       col = colors,
       pch = 19,
       main = "Scatterplot Colored by Group",
       xlab = "X",
       ylab = "Y")

  legend("topright",
         legend = c("Group 1", "Group 2"),
         col = c("blue", "orange"),
         pch = 19)
}, error = function(e) {
  message("Error in color-based scatterplot: ", e$message)
})
````

**Explanation:**

* `ifelse(...)` is used to assign color based on group membership.
* The third variable (`group`) is represented using color.
* A `legend()` is included for interpretation.

---

### Example 3: Bubble Plot (Size Encodes a Fourth Variable)

````r
# Bubble plot where size of point represents another variable
tryCatch({
  x <- runif(30, 1, 10)
  y <- runif(30, 1, 10)
  z <- runif(30, 5, 20)  # Bubble sizes

  plot(x, y,
       main = "Bubble Plot: Size ~ Z",
       xlab = "X-axis",
       ylab = "Y-axis",
       pch = 21,
       bg = "lightblue",
       cex = z / max(z))  # Normalize size
}, error = function(e) {
  message("Error in bubble plot: ", e$message)
})
````

**Explanation:**

* `pch = 21` gives filled circles.
* `bg = "lightblue"` fills the bubbles.
* `cex = z / max(z)` adjusts size based on the `z` variable.

---

### Example 4: RGB and Hex Colors

````r
# Using custom RGB color and hex codes
tryCatch({
  barplot(c(4, 6, 9),
          col = c(rgb(0.2, 0.5, 0.8), "#FF5733", "green"),
          names.arg = c("A", "B", "C"),
          main = "Barplot with RGB and Hex Colors")
}, error = function(e) {
  message("Error using RGB/Hex colors: ", e$message)
})
````

**Explanation:**

* `rgb(r, g, b)` allows you to specify exact color tones.
* Hex values like `#FF5733` are widely used in web design and accepted in R graphics.

---

### Best Practices for Multi-Dimensional Visualization

* Use **color** meaningfully, especially to represent categories or groups.
* Don't rely on color alone; combine with **shape** or **labels** when possible.
* Normalize size variables (as in `z / max(z)`) to maintain visual clarity.
* Always include a **legend** when encoding more than 2 variables.
