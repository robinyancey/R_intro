## Solutions (Part One)

### ✔️ **Solutions to Exercise 1**

```r
scores <- c(75, 80, 82, 90, 70)

mean_score <- mean(scores)      # 79.4
sd_score <- sd(scores)          # ≈ 7.66
max_score <- max(scores)        # 90

mean_score
sd_score
max_score
```

### ✔️ **Solutions to Exercise 2**

```r
mat <- matrix(1:6, nrow = 2, byrow = TRUE)
#      [,1] [,2] [,3]
# [1,]    1    2    3
# [2,]    4    5    6

value_23 <- mat[2, 3]           # 6
mat_trans <- t(mat)             # Transposed
row1_sum <- sum(mat[1, ])       # 1 + 2 + 3 = 6
print(mat)
value_23
mat_trans
row1_sum
```

## ✔️ Solutions to Exercise 4

### ✔️ Solutions to Exercise 5

```r
# Solution for Exercise 9

temperatures <- c(23.4, 21.8, 24.6, 22.0, 20.9, 25.5, 23.0)

# 1. Calculate the mean temperature
mean_temp <- mean(temperatures)

# 2. Calculate the median temperature
median_temp <- median(temperatures)

# 3. Find the maximum and minimum temperature
max_temp <- max(temperatures)
min_temp <- min(temperatures)

# 4. Find the range (difference between max and min)
range_temp <- max_temp - min_temp

# 5. Calculate the variance
variance_temp <- var(temperatures)

# Output to verify
mean_temp
median_temp
max_temp
min_temp
range_temp
variance_temp
```
