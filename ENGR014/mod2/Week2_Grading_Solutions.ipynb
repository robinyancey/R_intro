### **Solution 1 – Matrix Multiplication**

```
A <- matrix(c(1, 2, 3, 4, 5, 6), nrow = 2, byrow = TRUE)
B <- matrix(c(1, 2, 3, 4, 5, 6), nrow = 3, byrow = TRUE)
result <- A %*% B
print(result)
# Output: 2x2 matrix
```

### **Solution 2 – Handling NA and NULL**

```
x <- c(4, NA, 7, NULL, 3, NA)
x_clean <- na.omit(x)
mean_x <- mean(x_clean)
print(mean_x)
# Output: 4.666667
```

### **Solution 3 – List and Element Access**

```
person <- list(Name = "Alice", Age = 25, Grades = c(88, 92, 95))
second_grade <- person$Grades[2]
print(second_grade)
# Output: 92
```

### **Solution 4 – Class and Type Conversion**

```
num <- "42"
class(num)   # 'character'
num_numeric <- as.numeric(num)
result <- num_numeric * 2
print(result)
# Output: 84
```

### **Solution 5 – Normal Distribution Probability**

```
prob <- pnorm(80, mean = 70, sd = 10)
print(prob)
# Output: 0.8413447
```
