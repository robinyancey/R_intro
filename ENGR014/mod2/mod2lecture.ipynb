##  **2.0 Matrix Operations in R**

###  What is a Matrix?

A **matrix** in R is a two-dimensional data structure where:

* All elements must be of the **same data type**
* It's organized by **rows** and **columns**

Matrices are widely used in data science, machine learning, and statistical analysis for handling numeric datasets and performing **linear algebra operations**.

---

###  **Creating a Matrix in R**
>

````r
# Creating a matrix with numbers 1 to 9
my_matrix <- matrix(1:9, nrow = 3, ncol = 3)
print(my_matrix)
````


> The `matrix()` function fills by column by default.

````r
# Fill by row
matrix(1:6, nrow = 2, ncol = 3, byrow = TRUE)
````

###  **Basic Matrix Operations**

#### 1. **Matrix Addition and Subtraction**

Matrices can be added or subtracted if they are of the same dimension.

````r
A <- matrix(c(1, 2, 3, 4), nrow = 2)
B <- matrix(c(5, 6, 7, 8), nrow = 2)

A + B  # Addition
A - B  # Subtraction
````

#### 2. **Element-wise Multiplication**

This multiplies each corresponding element:

````r
A * B
````

#### 3. **Matrix Multiplication**

For actual **matrix multiplication** (dot product), use the `%*%` operator.

###Rule of Matrix Multiplication
####Let:

* Matrix A be of size m x n

* Matrix B be of size n x p

Then, A x B is possible, and the result is a matrix of size m x p.

The number of columns of A must equal the number of rows of B.

###Example 1: Multiply 2x3 with 3x2 Matrix
**Step 1: Create matrices A and B**

````r
# Matrix A (2x3)
A <- matrix(c(1, 2, 3, 4, 5, 6), nrow=2, byrow=TRUE)
# Matrix B (3x2)
B <- matrix(c(7, 8, 9, 10, 11, 12), nrow=3, byrow=TRUE)
````

````r
A
````

````r
B
````

### **Step 2: Matrix Multiply them**

````r
C <- A %*% B
print(C)
````

### **How the result is calculated**

Each element in result matrix **C\[i,j]** is computed as:

```
C[i,j] = A[i,1]*B[1,j] + A[i,2]*B[2,j] + A[i,3]*B[3,j]
```

#### For C\[1,1]:

```
= (1*7) + (2*9) + (3*11)
= 7 + 18 + 33 = 58
```

#### For C\[1,2]:

```
= (1*8) + (2*10) + (3*12)
= 8 + 20 + 36 = 64
```

#### For C\[2,1]:

```
= (4*7) + (5*9) + (6*11)
= 28 + 45 + 66 = 139
```

#### For C\[2,2]:

```
= (4*8) + (5*10) + (6*12)
= 32 + 50 + 72 = 154
```

````r
# Trying to multiply 2x3 with 2x2 (invalid)
A %*% matrix(c(1,2,3,4), nrow=2)
# Error: non-conformable arguments
````

### **Matrix Determinant and Inverse**
##  What is the **Determinant**?

The **determinant** is a **special number** that tells us about a matrix:

* If it's **not 0**, the matrix has an inverse.
* If it's **0**, the matrix **does NOT** have an inverse.

For a 2x2 matrix:

```
| a  b |
| c  d |
```

The determinant is calculated as:
**`(a*d) - (b*c)`**

### R Example:

````r
m <- matrix(c(2, 1, 3, 4), nrow = 2, byrow = TRUE)
print(m)
det(m)  # m is our matrix
````

The determinant is:

```
(2 * 4) - (3 * 1) = 8 - 3 = 5
```

##  What is the **Inverse of a Matrix**?

A matrix **inverse** is like dividing numbers ‚Äî it **"undoes"** the matrix.
If you multiply a matrix by its inverse, you get the **identity matrix**:

```
| 1 0 |
| 0 1 |
```

You can only find the inverse if the **determinant is not zero**.

###  R Example

````r
inverse_m <- solve(m)
print(inverse_m)
````

This gives the inverse of the matrix `m`.


It can also be calculated by swapping positions of a and d, putting negatives in front of b and c, and dividing everything by the determinant.

##  Verify: Multiply Matrix and Its Inverse

This will return the identity matrix:

````r
result <- m %*% inverse_m
print(result)
````

### In R, you can calculate the inverse of a matrix using the solve() function.
```r
solve(x)
```
**Example:**

````r
# Define a 2x2 matrix
A <- matrix(c(4, 7, 2, 6), nrow = 2, byrow = TRUE)

# Calculate the inverse
A_inv <- solve(A)

# Print the result
print(A_inv)
````

###Notes:
* The matrix must be square (same number of rows and columns).

* The matrix must be non-singular, meaning its determinant is not zero. You can check this with det(A).

###Handling Systems of Equations:
If you're solving a linear system Ax = b, you can use:

```r
x <- solve(A, b)
```

### This avoids explicitly calculating the inverse, which is more efficient and numerically stable.

### Full Code Summary:

````r
# Create a matrix
m <- matrix(c(2, 1, 3, 4), nrow = 2, byrow = TRUE)

# Print the matrix
print("Original matrix:")
print(m)

# Calculate determinant
d <- det(m)
print(paste("Determinant:", d))

# Check if determinant is not zero
if (d != 0) {
  # Calculate inverse
  inverse_m <- solve(m)
  print("Inverse of matrix:")
  print(inverse_m)

  # Verify identity matrix
  identity <- m %*% inverse_m
  print("Matrix multiplied by its inverse (Identity Matrix):")
  print(identity)
} else {
  print("Matrix has no inverse (determinant is 0)")
}
````

> NB: Only **square matrices** with non-zero determinants can be inverted.

# 2.3 Special Values, Classes, and Coercion in R

Understanding how R handles **special values**, **data types (classes)**, and **automatic or explicit coercion** is crucial for writing bug-free code and interpreting your results correctly.

---

### Special Values in R

R has built-in representations for unusual data points or errors. Here are the most common:

#### 1. `NA` ‚Äì Missing Values

Represents a missing value in your dataset.

````r
x <- c(1, 2, NA, 4)
mean(x)        # Returns NA
mean(x, na.rm = TRUE)  # Returns 2.33
````

#### 2. `NaN` ‚Äì Not a Number

Appears when a calculation is undefined (e.g., 0/0).

````r
0 / 0           # Returns NaN
is.nan(0 / 0)   # TRUE
````

#### 3. `Inf` and `-Inf` ‚Äì Infinity

Results from operations like dividing by zero.


````r
1 / 0       # Inf
-1 / 0      # -Inf
is.infinite(1 / 0)  # TRUE
````

#### 4. `NULL` ‚Äì Absence of a Value/Object

Unlike `NA`, `NULL` means the object doesn‚Äôt exist at all.


````r
x <- NULL
length(x)    # 0
````

### Data Classes in R

Everything in R has a **class** (or type). Common base types include:

| Type         | Example             |
| ------------ | ------------------- |
| `numeric`    | 1.5, 100            |
| `integer`    | 1L, 23L             |
| `character`  | "text", "R"         |
| `logical`    | TRUE, FALSE         |
| `complex`    | 1+2i                |
| `factor`     | Categorical values  |
| `list`       | Mixed-type elements |
| `data.frame` | Tabular structure   |

#### Checking Class

````r
x <- "Hello"
class(x)        # "character"

y <- 10L
class(y)        # "integer"
````

### Coercion in R

**Coercion** refers to converting one data type into another. R does this in two ways:

#### Implicit Coercion (automatic)

When combining different types in the same structure, R converts them to the **most flexible type** (usually character).

````r
c(1, "a")        # "1", "a" ‚Üí numeric is coerced to character
c(TRUE, 1)       # 1, 1     ‚Üí logical to numeric
````

#### Explicit Coercion (using functions)

R provides functions to force data into a specific type:

````r
as.numeric("5")         # 5
as.character(123)       # "123"
as.logical(0)           # FALSE
as.integer(3.5)         # 3 (truncates)
````


### Handling Coercion Errors

Not all coercions are valid:


````r
as.numeric("hello")     # NA with warning
as.logical("TRUE")      # NA
````

Always check your inputs and use `is.na()` to find failed conversions.

---

### Summary

* Use `is.na()`, `is.nan()`, `is.infinite()`, and `is.null()` to detect special values.
* `class()` helps you inspect the type of your data.
* Use `as.<type>()` for safe, explicit coercion.
* Coercion happens **silently** in many operations‚Äîunderstand the rules to avoid bugs.

---

### Practice Code

````r
# Special values
x <- c(2, NA, 4, NaN, Inf, -Inf, 8)
summary(x)

# Checking types
a <- 10
b <- "data"
c <- TRUE

class(a)
class(b)
class(c)

# Coercion
vec1 <- c(1, "two", TRUE)
class(vec1)   # Character

as.numeric("100")
as.logical(1)
as.integer(4.9)
````

# 2.4 Elementary Statistics in R

This lesson introduces you to the fundamental statistical tools available in R. These are essential for analyzing, summarizing, and interpreting data.

---

### Descriptive Statistics

#### 1. Mean (Average)

````r
x <- c(10, 20, 30, 40)
mean(x)  # Returns 25
````

#### 2. Median (Middle Value)

````r
median(x)  # Returns 25
````

The median is especially useful when your data contains outliers.

#### 3. Mode (Most Frequent Value)

R does **not** have a built-in `mode()` function in the statistical sense (i.e., most frequent value), so you can create one:

````r
get_mode <- function(v) {
  uniqv <- unique(v)
  uniqv[which.max(tabulate(match(v, uniqv)))]
}

x <- c(1, 2, 2, 3, 3, 3, 4)
get_mode(x)  # Returns 3
````

#### 4. Range (Minimum and Maximum)

````r
range(x)       # Returns c(min, max)
diff(range(x)) # Range spread
````

#### 5. Variance and Standard Deviation

* **Variance** measures the average squared deviation from the mean.
* **Standard deviation** is the square root of variance.

````r
var(x)  # Variance
sd(x)   # Standard deviation
````

#### 6. Quantiles

````r
quantile(x)          # Default 0%, 25%, 50%, 75%, 100%
quantile(x, probs = c(0.25, 0.75))  # Q1 and Q3
````

### Frequency Tables

You can create frequency tables for categorical data:

To examine relationships between two variables:

````r
x <- c(1, 2, 3, 4)
y <- c(2, 4, 6, 8)

cov(x, y)  # Covariance
cor(x, y)  # Correlation coefficient
````

Correlation ranges from -1 to +1 and shows direction and strength of the linear relationship.

# 2.5 Probability in R

Probability is the foundation of statistics. In this section, you'll learn how to simulate and work with basic probability concepts using R. This includes theoretical probability, random events, and commonly used distributions.

---

### What is Probability?

Probability measures the likelihood that a certain event will occur. It ranges from 0 (impossible) to 1 (certain).

For example:

* The probability of rolling a 3 on a fair 6-sided die is $\frac{1}{6} \approx 0.167$
* The probability of drawing an ace from a standard deck is $\frac{4}{52} = 0.077$

---

### Simulating Random Experiments in R

#### 1. Sample Space: `sample()`

The `sample()` function is used to simulate random draws.

````r
sample(1:6, size = 1)   # Simulate one die roll
sample(c("H", "T"), size = 10, replace = TRUE)  # Tossing a coin 10 times
````

* `replace = TRUE` means sampling **with replacement** (e.g., like tossing a coin multiple times).

#### 2. Setting a Random Seed

To make your simulation **reproducible**, use:


````r
#set.seed(125)
sample(1:10, 5)
````

### Theoretical Probability Distributions

R provides built-in support for many **probability distributions**, such as:

| Distribution | Description              | Common Function Prefix       |
| ------------ | ------------------------ | ---------------------------- |
| Binomial     | # of successes in trials | `dbinom`, `pbinom`, `rbinom` |
| Normal       | Bell curve               | `dnorm`, `pnorm`, `rnorm`    |
| Poisson      | Event counts             | `dpois`, `ppois`, `rpois`    |
| Uniform      | Equal likelihood         | `runif`                      |

The prefixes:

* `d` = density or probability
* `p` = cumulative probability
* `r` = random generation

# 2.6 Common Probability Distributions in R

Understanding probability distributions is key in statistics and data analysis. A probability distribution describes how values of a random variable are expected to behave. R provides extensive support for working with various distributions using consistent naming conventions for related functions.

---

### Overview of Common Distributions

Here are the most widely used distributions in R and their primary use cases:

| Distribution    | Description                            | Use Case Example                         |
| --------------- | -------------------------------------- | ---------------------------------------- |
| **Uniform**     | Equal probability for all values       | Random selection (e.g., lottery)         |
| **Binomial**    | Number of successes in `n` trials      | Coin tosses, yes/no surveys              |
| **Poisson**     | Count of events in fixed time/space    | Call center arrivals, website visits     |
| **Normal**      | Bell-shaped continuous distribution    | Heights, test scores, measurement errors |
| **Exponential** | Time between events in Poisson process | Time until failure, customer arrivals    |
| **Geometric**   | Trials until first success             | First success in repeated attempts       |

---

### Function Naming Convention in R

Each distribution has a family of functions that begin with a letter:

* `d` = Density (PDF/PMF)
* `p` = Cumulative distribution (CDF)
* `q` = Quantile function (inverse of CDF)
* `r` = Random value generation

Example:
For the Normal distribution: `dnorm()`, `pnorm()`, `qnorm()`, `rnorm()`
For the Binomial distribution: `dbinom()`, `pbinom()`, `qbinom()`, `rbinom()`

---

### 1. Uniform Distribution

All values are equally likely in a range.

````r
# Random sample from uniform distribution between 1 and 100
runif(5, min = 1, max = 100)
````

### 2. Binomial Distribution

Used when there are only two outcomes (success/failure) in a fixed number of trials.

````r
# Probability of 3 heads in 5 coin flips (0.5 chance of heads)
dbinom(3, size = 5, prob = 0.5)

# Simulate 10 coin toss experiments
rbinom(10, size = 5, prob = 0.5)
````

### 3. Poisson Distribution

Used for count data (e.g., number of emails received in an hour).

````r
# Probability of getting exactly 2 events when lambda = 4
dpois(2, lambda = 4)

# Simulate 10 random counts
rpois(10, lambda = 4)
````

### 4. Normal Distribution

Most common continuous distribution, symmetrical bell shape.

````r
# Generate and plot normal distribution
x <- rnorm(1000, mean = 50, sd = 10)
hist(x, breaks = 20, col = "lightblue", main = "Normal Distribution")
````

### 5. Exponential Distribution

Used for time between events (e.g., time until next customer enters).

````r
# Random times between events
rexp(5, rate = 1/5)  # mean = 5
````

### 6. Geometric Distribution

Number of failures before the first success.

````r
# Probability of 3 failures before first success (p = 0.3)
dgeom(3, prob = 0.3)

# Simulate trials
rgeom(5, prob = 0.3)
````

### Visualization Example

Visualize the Binomial Distribution:

````r
x <- 0:10
probs <- dbinom(x, size = 10, prob = 0.5)
barplot(probs, names.arg = x, col = "lightgreen", main = "Binomial Distribution P(X = x)")
````
